```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
hist2d = [ 0 for i in range(5000) ]
with open('./g_hist2d.txt', 'r') as f:
    for i, l in enumerate(f.readlines()):
        hist2d[i] = list( map(int, l.split(',')) )
```

```python
X_range = len(hist2d)
Y_range = len(hist2d[0])

X = list(range(X_range))
Y = list(range(Y_range))


COLOR_HIST = [ [0 for y in Y] for x in X ]

for i, line in enumerate(COLOR_HIST):
    for j, col in enumerate(line):
        COLOR_HIST[i][j] = hist2d[i][j]
```

```python
fig, ax = plt.subplots( 1,1, figsize=(150,100) )

ax.imshow( np.array(hist2d).T, cmap='hot')
```

```python
# Gibbs
def gibbs(g, a=a,b=b,c=c):
    return -a*g + b*(g**(2.0/3)) + c*(g**(-1.0/3))
```
